```python
class SmartDevice:
    def __init__(self, device_id, name):
        if device_id == "":
            raise ValueError("Device ID cannot be empty")
        self.name = name
        self.__device_id = device_id
        self.__power_status = "OFF"

    def turn_on(self):
        if self.__power_status == "ON":
            print(f"{self.name} is already ON")
        else:
            self.__power_status = "ON"
            print(f"{self.name} is now ON")

    def turn_off(self):
        if self.__power_status == "OFF":
            print(f"{self.name} is already OFF")
        else:
            self.__power_status = "OFF"
            print(f"{self.name} is now OFF")

    def display_info(self):
        print(f"Name: {self.name}")
        print(f"Device ID: {self.__device_id}")
        print(f"Power Status: {self.__power_status}")

    @property
    def device_id(self):
        return self.__device_id

    @property
    def power_status(self):
        return self.__power_status
```

```python
class TemperatureSensor(SmartDevice):
    def __init__(self, device_id, name, temperature):
        super().__init__(device_id, name)
        self.temperature = temperature

    def read_temperature(self):
        if self.power_status == "ON":
            print(f"The temperature on {self.name} is {self.temperature}°C.")
        else:
            print(f"{self.name} is OFF. Turn it ON to read the temperature.")

    def display_info(self):
        super().display_info()
        print(f"Temperature: {self.temperature}°C")
```

```python
class SmartLight(SmartDevice):
    def __init__(self, device_id, name, brightness):
        super().__init__(device_id, name)
        if brightness < 0 or brightness > 100:
            raise ValueError("Brightness must be between 0 and 100")
        self.brightness = brightness

    def increase_brightness(self):
        if self.brightness < 100:
            self.brightness += 10
            print(f"Brightness is now {self.brightness}%")
        else:
            print("Brightness is already at maximum.")

    def decrease_brightness(self):
        if self.brightness > 0:
            self.brightness -= 10
            print(f"Brightness is now {self.brightness}%")
        else:
            print("Brightness is already at minimum.")

    def display_info(self):
        super().display_info()
        print(f"Brightness: {self.brightness}%")
```

```python
class SecurityCamera(SmartDevice):
    def __init__(self, device_id, name):
        super().__init__(device_id, name)
        self.recording_status = "Not Recording"

    def start_recording(self):
        if self.power_status == "OFF":
            print(f"{self.name} is OFF. Turn it ON to start recording.")
        elif self.recording_status == "Recording":
            print(f"{self.name} is already recording.")
        else:
            self.recording_status = "Recording"
            print(f"{self.name} has started recording.")

    def stop_recording(self):
        if self.power_status == "OFF":
            print(f"{self.name} is OFF.")
        elif self.recording_status == "Not Recording":
            print(f"{self.name} is not recording.")
        else:
            self.recording_status = "Not Recording"
            print(f"{self.name} has stopped recording.")

    def display_info(self):
        super().display_info()
        print(f"Recording Status: {self.recording_status}")
```

```python
# OBJECT CREATION
temperature_sensor = TemperatureSensor("TS002", "Living Room", 20)
smart_light = SmartLight("SL002", "Bed Rooom Light", 50)
security_camera = SecurityCamera("SC001", "Front Door Camera")
```

```python
# MENU DRIVEN PROGRAM
while True:
    print("\n========== SMART DEVICE MANAGEMENT SYSTEM ==========")
    print("1. Display Device Information")
    print("2. Turn Device On")
    print("3. Turn Device Off")
    print("4. Read Temperature")
    print("5. Adjust Brightness")
    print("6. Start Recording")
    print("7. Stop Recording")
    print("8. Exit")

    user_choice = input("Enter your choice: ")

    if user_choice == "1":
        print("\n----- DEVICE INFORMATION -----")
        temperature_sensor.display_info()
        print()
        smart_light.display_info()
        print()
        security_camera.display_info()

    elif user_choice == "2":
        print("\nSelect Device")
        print("1. Temperature Sensor")
        print("2. Smart Light")
        print("3. Security Camera")

        device = input("Choice: ")

        if device == "1":
            temperature_sensor.turn_on()
        elif device == "2":
            smart_light.turn_on()
        elif device == "3":
            security_camera.turn_on()
        else:
            print("Invalid choice.")

    elif user_choice == "3":
        print("\nSelect Device")
        print("1. Temperature Sensor")
        print("2. Smart Light")
        print("3. Security Camera")

        device = input("Choice: ")

        if device == "1":
            temperature_sensor.turn_off()
        elif device == "2":
            smart_light.turn_off()
        elif device == "3":
            security_camera.turn_off()
        else:
            print("Invalid choice.")

    elif user_choice == "4":
        temperature_sensor.read_temperature()

    elif user_choice == "5":
        print("\nBrightness Control")
        print("1. Increase Brightness")
        print("2. Decrease Brightness")

        brightness = input("Choice: ")

        if brightness == "1":
            smart_light.increase_brightness()
        elif brightness == "2":
            smart_light.decrease_brightness()
        else:
            print("Invalid choice.")

    elif user_choice == "6":
        security_camera.start_recording()

    elif user_choice == "7":
        security_camera.stop_recording()

    elif user_choice == "8":
        print("\nExiting Smart Device Management System...")
        print("Goodbye!")
        break

    else:
        print("Invalid option. Please try again.")
```
